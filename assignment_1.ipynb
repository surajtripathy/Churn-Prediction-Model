```python
#Importing data from csv to dataframe.
import pandas as pd
data_file = pd.read_csv('Data/Churn_Modelling.csv')
data_file.head()
```

```python
#Since we won't need RowNumber, CustomerId and Surname in building our model, dropping it from DF.
data_file = data_file.drop(['RowNumber','CustomerId','Surname'],axis=1)
data_file.head()
```

```python
print(len(set(data_file['Geography'])))    #Finding number of unique elements in Geography tag
print(len(set(data_file['Gender'])))       #Finding number of unique elements in Gender tag

#Replacing Strings to Numbers. France:0 Spain:1 Germany:2
data_file['Geography'].replace('France',0,inplace=True)
data_file['Geography'].replace('Spain',1,inplace=True)
data_file['Geography'].replace('Germany',2,inplace=True)
#Replacing Strings to Numbers. Male:0 Female:1
data_file['Gender'].replace('Male',0,inplace=True)
data_file['Gender'].replace('Female',1,inplace=True)
```

```python
#After converting all into numbers
data_file.head()
```

```python
#Splitting total DF to training and test datasets. 
train_set = data_file.sample(frac=0.8,random_state=200)
test_set = data_file.drop(train_set.index)
```

```python
train_set.head()
```

```python
test_set.head()
```

```python
#As our targets as in 'Exited', we put them in the Y set of train and test sets.
Y_train = train_set['Exited']
X_train = train_set.drop('Exited',axis = 1)
Y_test = test_set['Exited']
X_test = test_set.drop('Exited',axis = 1)
```

```python
#Building a linear_reg model and testing it's accuracy.
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train, Y_train)
Y_pred = linreg.predict(X_test)
print("accuracy:", round(linreg.score(X_test, Y_test), 3))
```

```python
#Building a linear_reg model and testing it's accuracy.
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state=0)
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
print("accuracy:", round(logreg.score(X_test, Y_test), 3))
```
